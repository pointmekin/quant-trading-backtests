### MACD Signal

```python
import numpy as np

backcandles = 50

# calcualte macd signal
macd_signal = [0]*len(df)

for row in range(1, len(df)):
    if row < backcandles:
        continue
    # linear fit the last 15 data points of MACD and if the slope is positive using linear regression
    previous_macd = df.MACD[row-backcandles:row]
    previous_macd_signal = df.MACD_HIST[row-backcandles:row]
    previous_macd_slope = np.polyfit(range(0, len(previous_macd)), previous_macd, 1)[0]

    if(
        # df.MACD_SIGNAL[row]>df.MACD_SIGNAL[row-1] 
        df.MACD_SIGNAL[row] < 0  
        # and df.MACD[row] < 0
        # and df.ADX[row] > 25
        and previous_macd_slope < 0
    ):
        macd_signal[row]=1
    elif  (
        # df.MACD_SIGNAL[row]<df.MACD_SIGNAL[row-1] 
        df.MACD_SIGNAL[row] > 0 
        # and df.MACD[row] > 0
        # and df.ADX[row] > 25
        and previous_macd_slope > 0
    ):
        macd_signal[row]=-1

df['MACDSignal'] = macd_signal
```

### Bollinger band signal

```python
backcandles = 2

# calculate bollinger signal
bollinger_signal = [0]*len(df)

for row in range(1, len(df)):
  # if the past 14 candles are below the lower band and the current candle is above the lower band, then buy
  # if the past 14 candles are above the upper band and the current candle is below the upper band, then sell
  
  # pastCandles status
  pastCandles = df.iloc[row-backcandles:row-1]
  pastCandles_above_upperband = True
  for candle in pastCandles.iterrows():
    if candle[1].Close < candle[1]['BBU']:
      pastCandles_above_upperband = False
      break

  pastCandles_below_lowerband = True
  for candle in pastCandles.iterrows():
    if candle[1].Close < candle[1]['BBL']:
      pastCandles_below_lowerband = False
      break
  

  # current candle status
  currentcandle = df.iloc[row]
  currentcandle_above_upperband = currentcandle.Close > currentcandle['BBU']
  currentcandle_below_lowerband = currentcandle.Close < currentcandle['BBL']

  # buy
  if pastCandles_below_lowerband and currentcandle_above_upperband:
    bollinger_signal[row] = 1
  # sell
  elif pastCandles_above_upperband and currentcandle_below_lowerband:
    bollinger_signal[row] = -1
  else:
    bollinger_signal[row] = 0

df['BollingerBandSignal'] = bollinger_signal
```

### All signals

```python
def TotalSignal(l):
    if (
        df.BollingerBandSignal[l]==1
        and df.MACDSignal[l]==1
        # df.MACDSignal[l]==1
        # and df.RSI[l]>50
        # and df.RSI[l]<40
        # and df.Close[l] > df.MA_200[l]
    ):
            return 2
    if (
        df.BollingerBandSignal[l]==-1
        and df.MACDSignal[l]==- 1
        # df.MACDSignal[l]==- 1
        # and df.RSI[l]<50
        # and df.RSI[l]>60
        # and df.Close[l] < df.MA_200[l]
    ):
            return 1
    return 0
        
TotSignal = [0]*len(df)
for row in range(backcandles, len(df)): #careful backcandles used previous cell
    TotSignal[row] = TotalSignal(row)
df['TotalSignal'] = TotSignal
```

```python
from backtesting import Strategy
from backtesting import Backtest

class MyStrat(Strategy):
    initsize = 0.99
    mysize = initsize
    def init(self):
        super().init()
        self.signal1 = self.I(SIGNAL)

    def next(self):
        super().next()
        slatr = 1.5*self.data.ATR[-1]
        TPSLRatio = 1.75

        if len(self.trades)>0:
            if self.trades[-1].is_long and self.data.RSI[-1]>=90:
                self.trades[-1].close()
            elif self.trades[-1].is_short and self.data.RSI[-1]<=10:
                self.trades[-1].close()
        
        print("Datetime: ", self.data.index[-1])
        if self.signal1==2 and len(self.trades)==0:
            sl1 = self.data.Close[-1] - slatr
            tp1 = self.data.Close[-1] + slatr*TPSLRatio
            self.buy(sl=sl1, tp=tp1, size=self.mysize)
            print("Buy", self.data.Close[-1])
            print("Stop loss: ", sl1)
            print("Take profit: ", tp1)
            print("------")
        
        elif self.signal1==1 and len(self.trades)==0:         
            sl1 = self.data.Close[-1] + slatr
            tp1 = self.data.Close[-1] - slatr*TPSLRatio
            self.sell(sl=sl1, tp=tp1, size=self.mysize)
            print("Sell", self.data.Close[-1])
            print("Stop loss: ", sl1)
            print("Take profit: ", tp1)
            print("------")

        else:
            print("No signal")

bt = Backtest(dfpl, MyStrat, cash=100, margin=1/10, commission=0.00)
stat = bt.run()
```
