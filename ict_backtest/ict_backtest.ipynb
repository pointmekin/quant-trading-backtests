### Check FVG

```python
# Define a function to check for the fair value gap
def check_fair_value_gap(range1, range2, range3):
  # Extract the high and low values of each range
  low1, high1 = range1
  low2, high2 = range2
  low3, high3 = range3

  isDeclining = False
  isIncreasing = False

  # declining if the highs of ranges are decreasing
  if high1 > high2 > high3 and low1 > low2 > low3:
    isDeclining = True
  # increasing if the highs of ranges are increasing
  elif high1 < high2 < high3 and low1 < low2 < low3:
    isIncreasing = True
  
  if isDeclining:
    if low1 < high2:
      high2 = low1
    if high3 > low2:
      low2 = high3

    if high2 > low2:
      return {
        "isFairValueGap": True,
        "direction": "declining",
        "fvgHigh": high2,
        "fvgLow": low2,
        "entry": low2,
        "stopLoss": high1,
        "takeProfit": low2 - riskToRewardRatio * (high1 - low2)
      }
    return {
      "isFairValueGap": False
    }
    
  elif isIncreasing:
    if high1 > low2:
      low2 = high1
    if low3 < high2:
      high2 = low3

    if high2 > low2:
      return {
        "isFairValueGap": True,
        "direction": "increasing",
        "fvgHigh": high2,
        "fvgLow": low2,
        "entry": high2,
        "stopLoss": low1,
        "takeProfit": high2 + riskToRewardRatio * (high2 - low1)
      }
    return {
      "isFairValueGap": False
    }
  else :
    return {
      "isFairValueGap": False
    }

# print(check_fair_value_gap((10,20), (18,25), (23, 30)))

```

```python
import numpy as np

backcandles = 3

# calcualte macd signal
fvg_signal = [0]*len(df)
fvg_setup = [{
      "isFairValueGap": False
    }]*len(df)

hasSetup = False

for row in range(1, len(df)):
    # check if the time is between 2pm - 3pm GMT
    if df.index[row].hour == 14 and df.index[row].minute == 0:
        hasSetup = False
    if df.index[row].hour < 14 or df.index[row].hour > 15:
        continue

    # get the previous 3 candles
    prev_candles = df.iloc[row-backcandles + 1:row + 1]
    first_candle = prev_candles.iloc[0]
    middle_candle = prev_candles.iloc[1]
    last_candle = prev_candles.iloc[2]

    # check if the middle candle has fair value gap, meaning there are no wicks in that range from the first candle and last candle
    fvg = check_fair_value_gap(
        (first_candle.Low, first_candle.High),
        (middle_candle.Low, middle_candle.High),
        (last_candle.Low, last_candle.High)
    )
    if(
        fvg['isFairValueGap'] and
        fvg['direction'] == 'increasing' and
        not hasSetup
    ):
        hasSetup = True
        fvg_signal[row]=1
        fvg_setup[row]=fvg
    elif  (
        fvg['isFairValueGap'] and
        fvg['direction'] == 'declining' and
        not hasSetup
    ):
        hasSetup = True
        fvg_signal[row]=-1
        fvg_setup[row]=fvg

df['FVG_Signal'] = fvg_signal
df['FVG_Setup'] = fvg_setup
```
